## Load Data

```python
def extract_score(s):
    """Extract score from dictionary."""
    assert len(s) <= 1
    if 'scale_1' in s:
        return int(s['scale_1'])
    elif 'scale_2' in s:
        return int(s['scale_2'])
    elif 'scale_3' in s:
        return int(s['scale_3'])
    elif 'scale_4' in s:
        return int(s['scale_4'])
    elif 'scale_5' in s:
        return int(s['scale_5'])
    
    return np.nan

def get_similarity_scores_by_id(data_list, target_id):
    for item in data_list:
        if item['id'] == target_id:
            # Extract and return the two specific values
            return item['question_context_similarity'], item['question_response_similarity']
    return None, None  # Return None for both if no match is found

def load_json_data(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data

def load_results(filename, name, verbose=False):
    """Load results."""
    data = {}
    embedding_distance = load_json_data('embedding_distance.json')
    with open(filename, 'r') as f:
        for line in f:
            v = json.loads(line)

            if len(v['id']) > len('2000000'):
                if verbose:
                    print(v)
                continue

            h = hashlib.md5(v['displayed_text'].encode('utf-8')).hexdigest()
            context_similarity, response_similarity = get_similarity_scores_by_id(embedding_distance,int(v['id']))
            data[v['id']] = {
                'id': v['id'],
                'displayed_text': ' '.join(v['displayed_text'][:50].split()),
                #'span_annotations': v['span_annotations'],
                #'behavioral_data': v['behavioral_data'],

                'calculation_error': 1 if 'Calculation Error' in v['label_annotations'].get('Problems', '') else 0,

                'correctness_score': extract_score(v['label_annotations'].get('conclusion_Score', {})),
                'logic_score': extract_score(v['label_annotations'].get('logic_Score', {})),
                'truthfulness_score': extract_score(v['label_annotations'].get('truthfulness_Score', {})),

                'confidence_score': extract_score(v['label_annotations'].get('Confidence_Score', {})),

                'info': v['label_annotations'].get('textbox_input', {}).get('Descriptive grade', '') + \
                        ', ' + v['label_annotations'].get('textbox_input', {}).get('Comments', ''),
                'difficulty': int(v['id'][0]),  # 1=easy, 2=medium, 3=hard
                'originality': int(v['id'][1]),  # 0=copied, 1=paraphrased, 2=original
                'context': int(v['id'][2]),  # 0=no, 1=relevant, 2=vague, 3.irrelevant
                'author': int(v['id'][3]),  # 0=no, 1=relevant, 2=vague, 3.irrelevant
                'question_id': v['id'][-4:],
                'context_similarity':context_similarity,
                'response_similarity':response_similarity,
            }
    
    data = pd.DataFrame(data).T.set_index('id').sort_values(['question_id', 'context'])
    data['grader'] = name
    
    return data


# load
df = pd.concat([
    load_results('raw_data/abram.jsonl', name='abram'),
    load_results('raw_data/haoran.jsonl', name='haoran'),
    load_results('raw_data/louis.jsonl', name='louis'),
    load_results('raw_data/mikelixiang.jsonl', name='mikeli'),
    load_results('raw_data/Ryan_{{HYr}}.jsonl', name='ryan'),
    load_results('raw_data/ziwei.jsonl', name='ziwei'),
])

pd.set_option('display.max_rows', 200)
df
```

```python
# Function to normalize scores
def normalize_scores(scores):
    return (scores / 2.5) - 1
def calculate_statistics(grouped, bins):
    """Calculate mean and standard error for normalized scores within each bin of context similarity using pandas."""
    grouped['bin_index'] = pd.cut(grouped['context_similarity'], bins=bins, labels=range(len(bins)-1), right=True)
    binned_data = grouped.groupby('bin_index')['normalized_score']
    
    mean_scores = binned_data.mean()
    sem = binned_data.sem()

    return mean_scores, sem, binned_data.count()

# Define a function to extract benchmark IDs
def benchmark_id(qid):
    numx=int(qid)%10000
    
    numy=int(qid)//100000
    return int(numy*100000+numx)
def calculate_improvement_statistics(grouped, bins):
    grouped['bin_index'] = pd.cut(grouped['context_similarity'], bins=bins, labels=range(len(bins)-1), right=True)
    binned_data = grouped.groupby('bin_index')['score_improvement']
    
    mean_improvements = binned_data.mean()
    sem = binned_data.sem()
    counts = binned_data.count()
    
    return mean_improvements, sem, counts
# Group by 'id' and calculate mean for 'context_similarity' and each score
def prepare_data(type, number):
    filtered_df = df[(df['context'] != 0) & (df[type] == number)]
    filtered_df_2=df[(df['context'] == 0) & (df[type] == number)]
    grouped = filtered_df.groupby('id').agg({
        'context_similarity': 'mean',
        'correctness_score': 'mean',
        'logic_score': 'mean',
        'truthfulness_score': 'mean'
    }).reset_index()
    
    grouped_2 = filtered_df_2.groupby('id').agg({
        'context_similarity': 'mean',
        'correctness_score': 'mean',
        'logic_score': 'mean',
        'truthfulness_score': 'mean'
    }).reset_index()

    # Normalize mean scores
    grouped['normalized_correctness_score'] = normalize_scores(grouped['correctness_score'])
    grouped['normalized_logic_score'] = normalize_scores(grouped['logic_score'])
    grouped['normalized_truthfulness_score'] = normalize_scores(grouped['truthfulness_score'])
    
    grouped_2['normalized_correctness_score'] = normalize_scores(grouped_2['correctness_score'])
    grouped_2['normalized_logic_score'] = normalize_scores(grouped_2['logic_score'])
    grouped_2['normalized_truthfulness_score'] = normalize_scores(grouped_2['truthfulness_score'])
    grouped['normalized_score']=(grouped['normalized_correctness_score']+grouped['normalized_logic_score']+grouped['normalized_truthfulness_score'])/3
    grouped_2['normalized_score']=(grouped_2['normalized_correctness_score']+grouped_2['normalized_logic_score']+grouped_2['normalized_truthfulness_score'])/3
    # Normalize scores as provided
    grouped['normalized_score'] = (grouped['normalized_correctness_score'] + grouped['normalized_logic_score'] + grouped['normalized_truthfulness_score']) / 3
    grouped_2['normalized_score'] = (grouped_2['normalized_correctness_score'] + grouped_2['normalized_logic_score'] + grouped_2['normalized_truthfulness_score']) / 3
    # Apply the function to create a benchmark_id column in 'grouped'
    grouped['benchmark_id'] = grouped['id'].apply(benchmark_id)
    
    # Set the index of grouped_2 to 'id' and check data type consistency
    grouped_2.set_index('id', inplace=True)
    
    # Ensure the index of grouped_2 is an integer (if not already)
    grouped_2.index = grouped_2.index.astype(int)
    
    # Use map to pull the normalized scores from grouped_2 into grouped
    grouped['benchmark_score'] = grouped['benchmark_id'].map(grouped_2['normalized_score'])
    
    # Calculate the improvement
    grouped['score_improvement'] = grouped['normalized_score'] - grouped['benchmark_score']
    return grouped,grouped_2
```

```python
# Loop through all combinations of difficulty and originality
# Difficulty and Originality mappings
difficulty_mapping = {1: 'Easy', 2: 'Medium', 3: 'Hard'}
originality_mapping = {0: 'Copied', 1: 'Paraphrased', 2: 'Original'}
fig, axes = plt.subplots(2, 3, figsize=(18, 18), sharey=True)

# Define bins for all plots
bins = np.linspace(0, 0.8, num=5)
for i, difficulty in enumerate([1, 2, 3]):
    grouped,grouped_2=prepare_data('difficulty', difficulty)
    mean_improvements, sem, counts = calculate_improvement_statistics(grouped, bins)

    bin_midpoints = [(bins[k] + bins[k+1]) / 2 for k in range(len(bins) - 1)]
    bin_width = np.min(np.diff(bins)) * 0.8  # Bin width as 80% of the interval

    ax = axes[0, i]
    bars = ax.bar(bin_midpoints, mean_improvements, width=bin_width, yerr=sem, align='center', alpha=0.7, capsize=5, color='skyblue')
        
    # Print the size of each bin on top of each bar
    for bar, count in zip(bars, counts):
        height = bar.get_height()
        ax.annotate(f'{count}', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points", ha='center', va='bottom', fontsize=12)
    ax.set_title(f'Difficulty: {difficulty_mapping[difficulty]}', fontsize=14)
    ax.set_xlabel('Context Similarity Bins', fontsize=12)
    ax.set_ylabel('Mean Score Improvement', fontsize=12)
    ax.set_xticks(bin_midpoints)
    ax.set_xticklabels([f"{m:.2f}" for m in bin_midpoints], fontsize=14)
    ax.tick_params(axis='y', labelsize=14)
    ax.grid(True)
for j, originality in enumerate([0, 1, 2]):
    grouped,grouped_2=prepare_data('originality', originality)
    mean_improvements, sem, counts = calculate_improvement_statistics(grouped, bins)

    bin_midpoints = [(bins[k] + bins[k+1]) / 2 for k in range(len(bins) - 1)]
    bin_width = np.min(np.diff(bins)) * 0.8  # Bin width as 80% of the interval

    ax = axes[1, j]
    bars = ax.bar(bin_midpoints, mean_improvements, width=bin_width, yerr=sem, align='center', alpha=0.7, capsize=5, color='skyblue')
        
    # Print the size of each bin on top of each bar
    for bar, count in zip(bars, counts):
        height = bar.get_height()
        ax.annotate(f'{count}', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points", ha='center', va='bottom', fontsize=12)
    ax.set_title(f'Originality: {originality_mapping[originality]}', fontsize=14)
    ax.set_xlabel('Context Similarity Bins', fontsize=12)
    ax.set_ylabel('Mean Score Improvement', fontsize=12)
    ax.set_xticks(bin_midpoints)
    ax.set_xticklabels([f"{m:.2f}" for m in bin_midpoints], fontsize=14)
    ax.tick_params(axis='y', labelsize=14)
    ax.grid(True)

# Display the plot
plt.tight_layout()  # Adjust layout to make sure everything fits without overlap
plt.show()
```
